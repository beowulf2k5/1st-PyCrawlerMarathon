# 作業目標
請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
#1.這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
import requests
import json

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
popular = r.text

r2 = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
regular = r2.text


types = [popular, regular]


for i in types:
    data = json.loads(i)
    if i == types[0]:
        forum = '熱門文章'
    else:
        forum = '非熱門文章'      
    
    keylist = data[0].keys()
    
    print(forum + '回傳資料筆數 : ' +  str(len(data)))
    print('資料欄位:')
    for i in keylist:
        print(i)
    print('\n\n\n')


```

```python
#2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

data = json.loads(response)

for i in data:
    print('標題: ' + i.get('title') + '\n')
    print('貼文時間: ' + i.get('excerpt') + '\n')
    print('留言人數: ' + str(i.get('commentCount')) + '\n')
    print('按讚人數: ' + str(i.get('likeCount')) + '\n\n\n')
```

```python
#3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
popular = r.text

r2 = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
regular = r2.text


types = [popular, regular]


for i in types:
    totalComment = 0
    totalLike = 0
    data = json.loads(i)
    if i == types[0]:
        forum = '熱門文章'
    else:
        forum = '非熱門文章'
    
    for i in data:
        totalComment += i.get('commentCount')
        totalLike += i.get('likeCount')
    
    print(forum + '平均留言人數 : ' +  str(round(totalComment/len(data))))
    print(forum + '平均按讚人數 : ' +  str(round((totalLike/len(data)), 2)))
    print('\n\n\n')

    
```
