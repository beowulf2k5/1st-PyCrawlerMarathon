# ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup
import datetime
import re
import pandas as pd


today = datetime.date.today().strftime("%Y/%m/%d")

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html.parser")

alist = []

for d in soup.find(class_="part_list_2").find_all('h3'):
    if (d.find(class_="date").text)[:10] == today:
        print(d.find(class_="date").text, d.find('em').text, d.find_all('a')[-1].text)
        alist.append([d.find(class_="date").text, d.find('em').text, d.find_all('a')[-1].text])

```

# ② 如果想要依照類別分類，怎麼儲存會比較好？¶


```python
labels = ['Date', 'Category', 'Subject']
df = pd.DataFrame(alist, columns=labels)
df
```

# ③ 哪一個類別的文章最多¶


```python
temp = df.groupby('Category')
temp.size().sort_values(ascending = False)
```
